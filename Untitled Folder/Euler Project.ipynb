```python
def prob_one(n):
    num_list = []
    for i in range(1,n):
        if i % 3 == 0:
            num_list.append(i)
        elif i % 5 == 0:
            num_list.append(i)
    return sum(num_list)
```

```python
def is_prime(x):
    factors = []
    for i in range(2,x):
        if x % i == 0:
            factors.append(i)
        if factors == []:
            return True
        else:
            return False
```

```python
#Variables
rangeToCheck = range(100,1000)
biggestSoFar = 0

# Function to reverse strings
def reverseStr(string):
    return string[::-1]

# Function to check for palindromes
def palindromeCheck(possible):
    if possible == reverseStr(possible) and possible > str(biggestSoFar):
            return possible

# Run the sums
for i in rangeToCheck:
    j = i+1
    while j < 1000:
        n = i*j
        if str(n) == reverseStr(str(n)) and n > biggestSoFar:
            biggestSoFar = n
        j += 1

# Result
print('The largest palindrome is ' + str(biggestSoFar))
```

```python
n  n/2 (n is even)
n  3n + 1 (n is odd)
```

```python
def odd(n):
    return n * 3 + 1
    
def even(n):
    return n / 2
```

```python
counter = 0
def sequence(n):
    if n == 1:
        global counter
        return counter
    else:
        if n % 2 == 0:
            counter += 1
            new_n = even(n)
            sequence(new_n)
        else:
            counter +=1
            new_n = odd(n)
            sequence(new_n)
```

```python
sequence(5)
```

```python
sequence(1)
```

```python
lengths = []
numbers = []
for i in range(1, 1000001):
    counter = 0
    sequence(i)
    lengths.append(counter)
    numbers.append(i)
```
