## One

```python
def prob_one(n):
    num_list = []
    for i in range(1,n):
        if i % 3 == 0:
            num_list.append(i)
        elif i % 5 == 0:
            num_list.append(i)
    return sum(num_list)
```

```python
prob_one(10)
```

```python
prob_one(1000)
```

```python
def is_prime(x):
    factors = []
    for i in range(2,x):
        if x % i == 0:
            factors.append(i)
        if factors == []:
            return True
        else:
            return False
```

```python
def factors(n):
    results = set()
    for i in range(1, int(n**.5 + 1)):
        if n % i == 0:
            results.add(i)
            results.add(int(n/i))
    return results
```

```python
factors(60)
```

```python
factors(600851475143)
```

```python
for i in factors(600851475143):
    if factors(i) == {1, i}:
        print(i)
```

```python
def factor_pairs(n):
    results = []
    for i in range(1, int(n**.5 + 1)):
        pair = set()
        if n % i == 0:
            if len(str(i)) == 3:
                if len(str(int(n/i))) == 3:
                    pair.add(i)
                    pair.add(int(n/i))
                    results.append(pair)
    return results
```

```python
#Variables
rangeToCheck = range(100,1000)
biggestSoFar = 0

# Function to reverse strings
def reverseStr(string):
    return string[::-1]

# Function to check for palindromes
def palindromeCheck(possible):
    if possible == reverseStr(possible) and possible > str(biggestSoFar):
            return possible

# Run the sums
for i in rangeToCheck:
    j = i+1
    while j < 1000:
        n = i*j
        if str(n) == reverseStr(str(n)) and n > biggestSoFar:
            biggestSoFar = n
        j += 1

# Result
print('The largest palindrome is ' + str(biggestSoFar))
```
